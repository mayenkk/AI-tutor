```python
from typing import Literal
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.pydantic_v1 import BaseModel, Field
```

```python
from langchain_groq import ChatGroq
from google.colab import userdata
import os
groq_api_key = userdata.get('GROQ_API_KEY')
```

```python
llm = ChatGroq(groq_api_key = groq_api_key, model_name="Llama-3.1-70b-Versatile" )
```

```python
from typing import List
from typing_extensions import TypedDict

class GraphState(TypedDict):
  question: str
  answer: str
  user_level: int
  user_input: str
  end: bool
```

```python
def greeting(state: GraphState) -> GraphState:
    print("Hello! I am your AI tutor.")
    level = input("Please enter your level: beginner, amateur, or pro: ").lower()
    user_level = None

    if level == "beginner":
        user_level = 1
    elif level == "amateur":
        user_level = 5
    elif level == "pro":
        user_level = 8
    else:
        print("Invalid input, setting level to beginner.")
        user_level = 1

    # Store user level in the state
    state["user_level"] = user_level
    return state
```

```python
def generate_question_from_pdf(state: GraphState) -> GraphState:
    user_level = state.get("user_level")
    context = pdf_context  # Using the entire PDF as context

    if user_level is None:
        user_level = 1
        state['user_level'] = user_level

    # Use the context to generate the question
    prompt = f"You are a good AI agent. You will construct a good multi-choice question with the appropriate user_level from the given context. Construct a question from the following PDF context with a difficulty level of {user_level} (1-10): {context}. Just give the question, not the query or the answer."
    question = llm.invoke(prompt)

    state["question"] = question.content
    return state
```

```python
def evaluate_answer(state: GraphState) -> GraphState:
    question = state["question"]
    user_answer = input(f"{question}: ")
    context = pdf_context
    # Prompt to check if the answer is correct
    prompt = f"For the question '{question}' from the context '{context}', is the answer '{user_answer}' correct?"
    result = llm.invoke(prompt)

    if "correct" in result.content.lower():
        print("Correct answer!")
        state["user_level"] += 1  # Increase level
    else:
        print("Incorrect answer.")
        state["user_level"] -= 1  # Decrease level
        # Provide the correct answer
        correct_answer_prompt = f"You are an intelligent AI agent. You will answer the question '{question}' from the context '{context}' correctly, and will be awarded later."
        correct_answer = llm.invoke(correct_answer_prompt)
        print(f"The correct answer is: {correct_answer.content}")

    return state
```

```python
from langgraph.graph import END, StateGraph, START
def continue_or_exit(state: GraphState) -> GraphState:
    print("Do you want to continue? (yes/no)")
    user_input = input().strip().lower()
    if user_input == "no":
        state["end"] = True  # Set an end flag in the state
    elif user_input == "yes":
        state["end"] = False
    else:
        print("Invalid input. Please respond with yes or no.")
        return continue_or_exit(state)
    return state
```

```python
workflow = StateGraph(GraphState)

workflow.add_node("greeting", greeting)
workflow.add_node("generate_question_from_pdf", generate_question_from_pdf)
workflow.add_node("evaluate_answer", evaluate_answer)
workflow.add_node("continue_or_exit", continue_or_exit)

# Define edges for the conversation flow
workflow.add_edge(START, "greeting")
workflow.add_edge("greeting", "generate_question_from_pdf")
workflow.add_edge("generate_question_from_pdf", "evaluate_answer")
workflow.add_edge("evaluate_answer", "continue_or_exit")
workflow.add_conditional_edges(
    "continue_or_exit",
    lambda x: END if x.get("end", False) else "generate_question_from_pdf",
    {
        "generate_question_from_pdf": "generate_question_from_pdf",
        END: END
    }
)

# Compile the workflow
app = workflow.compile()
```

```python
initial_state = {
    "question": "",  # Initial question state
    "answer": "",    # Initial answer state
    "user_input": "",  # Input from user
    "user_level": None,  # Initial user level (set during interaction)
    "end": False
}

# Start the application with the initial state
app.invoke(initial_state)
```
